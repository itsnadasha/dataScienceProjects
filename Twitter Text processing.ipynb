## handeling texts - tweets

```python
import pandas as pd
import re

# ÿ™ÿπÿ±ŸäŸÅ ÿßŸÑÿØÿßŸÑÿ© ŸÑÿ•ÿ≤ÿßŸÑÿ© ÿßŸÑŸÉŸÑŸÖÿßÿ™ ÿßŸÑÿ™Ÿä ÿ™ÿ®ÿØÿ£ ÿ®ŸÄ "@"
#write a function that remove the @ in tweets, text column
def remove_mentions(text):
    return re.sub(r'@\w+', '', text).strip()


df['text'] = df['text'].apply(remove_mentions)
```

```python
import string # Import the string module to use string.punctuation

#this method removes the punctuation marks
def remove_punct(text):
  text_nopunct = "".join([char for char in text if char not in string.punctuation])
  return text_nopunct

df['text'] = df['text'].apply(lambda x: remove_punct(x))
df.tail(10)
```
